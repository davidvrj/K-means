Para mostrar y entender el proceso k-means usaremos una matriz de puntos de tamaño 300x2, donde cada fila proporciona las coordenadas (x, y) de un punto en un mapa. Haremos el diagrama de dispersión de estos puntos y usaremos el diagrama de dispersión para adivinar cuántos grupos hay.

**Paso 1:** Cargamos los datos e importamos las librerias a ocupar

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('ch1ex1.csv') 
points = df.values
```

**Paso 2**: Creamos una matriz llamada xs que contenga los valores de los puntos [:, 0], es decir, la columna 0 de los puntos:

```python
xs = points[:,0]
```

**Paso 3**: Creamos una matriz llamada ys que contenga los valores de los puntos [:, 1], es decir, la columna 1 de los puntos

```python
ys = points[:,1]
```

**Paso 4 :** Hacemos un diagrama de dispersión pasando xs e ys a la función plt.scatter () y mostramos el diagrama con la función .show()

```python
 plt.scatter(xs, ys)
 plt.show()
```

Desde el diagrama de dispersión del anterior, se puede ver que los puntos parecen separarse en 3 grupos. Ahora vamos a crear un modelo KMeans para encontrar 3 grupos y ajustarlo a los puntos de datos anteriores. Después de que el modelo se haya ajustado, obtendremos las etiquetas de clúster para puntos y también para algunos puntos nuevos utilizando el método .predict ().

**Paso 5** : Ocuparemos una matriz para los puntos nuevos que llamramemos new_points.

```python
new_df = pd.read_csv('ch1ex2.csv')
new_points = new_df.values
```

**Paso 6** : Importamos KMeans desde sklearn.cluster

```python
from sklearn.cluster import KMeans
```

**Paso 7**: Usando KMeans (), creamos una instancia de KMeans llamada modelo para encontrar 3 grupos. Para especificar el número de clústeres, use el argumento de palabra clave n_clusters

```python
model = KMeans(n_clusters=3)
```

**Paso 8:** 
Utilizamos el método de modelo .fit () para ajustar el modelo a la matriz de puntos "points".

```python
model.fit(points)
```

**Paso 9** : Imprimimos las etiquetas para observarlas

```python
labels = model.predict(points)
```

```python
print(labels)
```

**Paso 10**: Usamos el método de modelo .predict () para predecir las etiquetas de clúster de new_points, asignando el resultado a new_labels. Hay que tener en cuenta que KMeans puede asignar puntos nunca antes vistos a los grupos que ya hayamos encontrado.

```python
new_labels = model.predict(new_points)
```

**Paso 11:** Hacemos el diagrama de dispersión de xs y ys, especificando los argumentos de la palabra clave c = labels para colorear los puntos por su etiqueta de clúster. Veremos que KMeans identifica los grupos

```python
 plt.scatter(xs, ys, c=labels)
 plt.show()
```

Vemos que a simple vista, podemos identificar los clusters, pero en esta práctica también añadiremos los centroides de los clusters.

**Paso 12** : Obtenemos las coordenadas de los centroides utilizando el atributo .cluster_centers_ del modelo. Asignamos a ellos a "centroids".

```python
centroids = model.cluster_centers_
```

**Paso 13:** Asignamos la columna 0 de centroides a centroides_x y la columna 1 de centroides también a centroides_uy.

```python
centroids_x = centroids[:,0]
centroids_y = centroids[:,1]
```

**Paso 15:** En una sola celda, creamos dos diagramas de dispersión (esto mostrará los dos uno encima del otro). Llamamos a plt.show () solo una vez, al final.
En primer lugar, crea el diagrama de dispersión que hicimos arriba. En segundo lugar, hacemos un diagrama de dispersión de centroides_x y centroides_y, usando 'X' (una cruz) como marcador especificando el parámetro marcador. Establecemos el tamaño de los marcadores en 200 usando s = 200.

```python
plt.scatter(xs, ys, c=labels)
plt.scatter(centroids_x, centroids_y, marker='X', s=200)
plt.show()
```
